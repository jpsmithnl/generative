```python
import drawSvg as draw
import math
import random
random.seed(42)

d = draw.Drawing(400, 400, origin='center', displayInline=False)

def circle(n_points=40, radius=50, x_offset=0, y_offset=0, quadrant=0):
    points = []
    for i in range(n_points+1):
        theta = i*90./float(n_points)+90.*quadrant
        theta = math.radians(theta)
        x = radius * math.cos(theta) 
        y = radius * math.sin(theta)
        x += x_offset
        y += y_offset
        points.extend((x,y))
    return points

def draw_block(
    n_points=100,
    n_circles=20,
    spacing=10,
    block_size=200,
    quadrant=1,
    # x_offset=block_size,
    x_offset=200,
    y_offset=0,
    **line_config):
    
    for i in range(0,n_circles):
        radius = spacing * i
        circle_pts = circle(n_points, radius=radius, x_offset=x_offset, y_offset=y_offset, quadrant=quadrant) 
        if not circle_pts:
            continue
        d.append( draw.Lines(*circle_pts, **line_config))
        
line_config = dict(
    close=False,
    fill_opacity=0.0,
    stroke='orangered'
)

block_size=200
draw_block(
    quadrant=1,
    x_offset=block_size,
    y_offset=0,
    **line_config)
draw_block(
    quadrant=3,
    x_offset=0,
    y_offset=block_size,
    **line_config)
d
```

```python
import random
random.seed(5)
block_size=200

d = draw.Drawing(block_size*6, block_size*4, origin=(0,0), displayInline=False)

def arc(n_points=40, radius=50, x0=0, y0=0, w=200, h=200, quadrant=0):
    points = []
    for i in range(n_points+1):
        theta = i*90./float(n_points)+90.*quadrant
        theta = math.radians(theta)
        x = radius * math.cos(theta) + x0
        y = radius * math.sin(theta) + y0
        match quadrant:
            case 1:
                x += w
            case 2:
                x += w
                y += h
            case 3:
                y += h
        points.extend((x,y))
    d.append( draw.Lines(*points, **line_config))
        
    points=[]
    for i in range(n_points+1):
        theta = i*90./float(n_points)+90.*(quadrant+2)
        theta = math.radians(theta)
        x = radius * math.cos(theta) + x0
        y = radius * math.sin(theta) + y0
        match quadrant:
            case 0:
                x += w
                y += h
                r = math.sqrt((x0-x)**2 + (y0-y)**2)
            case 1:
                y += h
                r = math.sqrt((x0+w-x)**2 + (y0-y)**2)
            case 2:
                r = math.sqrt((x0+w-x)**2 + (y0+h-y)**2)
            case 3:
                x += w
                r = math.sqrt((x0-x)**2 + (y0+h-y)**2)
        if r > 200:
            points.extend((x,y))
        else:
            if points:
                d.append(draw.Lines(*points, **line_config))
            points = []
    if points:
        d.append(draw.Lines(*points, **line_config))

n_points=100
n_circles=20
spacing=10
radius=block_size


for x0 in range(0,4*block_size,block_size):
    for y0 in range(0,4*block_size,block_size):
        quadrant = random.choice([0,1,2,3])
        for i in range(0,n_circles+1):
            radius = spacing * i
            arc(n_points, radius=radius, x0=x0, y0=y0, w=block_size, h=block_size, quadrant=quadrant) 
d.savePng("parquet_spirals.png")
d
```
