```python
import drawSvg as draw
import math
import random
random.seed(42)

d = draw.Drawing(800, 400, origin='center', displayInline=False)

def circle(n_points=40, radius=50, x_offset=0, y_offset=0, quadrant=0):
    points = []
    for i in range(n_points+1):
        theta = i*90./float(n_points)+90.*quadrant
        theta = math.radians(theta)
        x = radius * math.cos(theta) 
        y = radius * math.sin(theta)
        x += x_offset
        y += y_offset
        points.extend((x,y))
    return points

def draw_block(
    n_points=100,
    n_circles=20,
    spacing=10,
    block_size=200,
    quadrant=1,
    x_offset=block_size,
    y_offset=0,
    **line_config):
    
    for i in range(0,n_circles):
        radius = spacing * i
        circle_pts = circle(n_points, radius=radius, x_offset=x_offset, y_offset=y_offset, quadrant=quadrant) 
        if not circle_pts:
            continue
        d.append( draw.Lines(*circle_pts, **line_config))
        
line_config = dict(
    close=False,
    fill_opacity=0.0,
    stroke='orangered'
)

draw_block(
    quadrant=1,
    x_offset=block_size,
    y_offset=0,
    **line_config)
draw_block(    
    quadrant=0,
    x_offset=-block_size,
    y_offset=0,
    **line_config)
draw_block(  
    quadrant=2,  
    x_offset=0,
    y_offset=0,
    **line_config)
draw_block( 
    quadrant=3,   
    x_offset=0,
    y_offset=0,
    **line_config)
draw_block( 
    quadrant=2,   
    x_offset=-block_size,
    y_offset=0,
    **line_config)
draw_block( 
    quadrant=1,   
    x_offset=-block_size,
    y_offset=0,
    **line_config)
draw_block( 
    quadrant=,   
    x_offset=2*block_size,
    y_offset=0,
    **line_config)
draw_block( 
    quadrant=1,   
    x_offset=2*block_size,
    y_offset=0,
    **line_config)
d
```
