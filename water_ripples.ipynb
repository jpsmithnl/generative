```python
import drawSvg as draw
import math
import random
random.seed(42)

d = draw.Drawing(400, 400, origin='center', displayInline=False)
b = draw.Drawing(400, 400, origin='center', displayInline=False)
r = draw.Drawing(400, 400, origin='center', displayInline=False)

def circle(offsets, n_points=40, radius=50, x0=0, y0=0):
    points = []
    for i,d in enumerate(list(offsets)+[offsets[0]]):
        theta = i*360./float(n_points)
        x = (radius+d) * math.cos(math.radians(theta))+x0
        y = (radius+d) * math.sin(math.radians(theta))+y0
        points.extend((x,y))
    return points

line_config = dict(
    close=False,
    fill_opacity=0.0,
    stroke='black'
)

n_points=100
increment=15
offsets = list([1]*n_points)
delta = 1
random_changes = list(random.uniform(-delta, delta) for o in offsets)
smoothed = []
window_size = 20
for i in range(len(random_changes)):
    sample = list(random_changes+random_changes[:window_size])[i: i + window_size]
    smoothed.append(sum(sample)/float(len(sample)))
    
for i in range(0,25):
    radius = increment*1.1**i
    offsetsa = list(a+(b*(2**i)*.000003) for a,b in zip(offsets, smoothed))
    d.append( draw.Lines(*circle(offsetsa,n_points,radius,0,0), **line_config))
    b.append( draw.Lines(*circle(offsetsa,n_points,radius,0,0), **line_config))

n_points=100
increment=5
offsets = list([.1]*n_points)
delta = 1
random_changes = list(random.uniform(-delta, delta) for o in offsets)
smoothed = []
window_size = 20

for i in range(len(random_changes)):
    sample = list(random_changes+random_changes[:window_size])[i: i + window_size]
    smoothed.append(sum(sample)/float(len(sample)))
for i in range(0,15):
    radius = increment*1.2**i
    offsetsa = list(a+(b*(2**i)*.001) for a,b in zip(offsets, smoothed))
    d.append( draw.Lines(*circle(offsetsa,n_points,radius,100,100), **line_config))
    b.append( draw.Lines(*circle(offsetsa,n_points,radius,100,100), **line_config))
    
n_points=100
increment=3
offsets = list([.1]*n_points)
delta = 1
random_changes = list(random.uniform(-delta, delta) for o in offsets)
smoothed = []
window_size = 20

for i in range(len(random_changes)):
    sample = list(random_changes+random_changes[:window_size])[i: i + window_size]
    smoothed.append(sum(sample)/float(len(sample)))
for i in range(0,10):
    radius = increment*1.3**i
    offsetsa = list(a+(b*(2**i)*.02) for a,b in zip(offsets, smoothed))
    d.append( draw.Lines(*circle(offsetsa,n_points,radius,60,-80), **line_config))
    b.append( draw.Lines(*circle(offsetsa,n_points,radius,60,-80), **line_config))
    
    
for x in range(10):
    for y in range(30):
        d.append(draw.Circle(-(x*7+5), y*7, 1, fill_opacity=0, stroke_width=1, stroke='red'))
        r.append(draw.Circle(-(x*7+5), y*7, 1, fill_opacity=0, stroke_width=1, stroke='red'))

b.saveSvg('blacks.svg')
r.saveSvg('reds.svg')
d.saveSvg('drawing.svg')
d.savePng("ripples.png")
d
```
