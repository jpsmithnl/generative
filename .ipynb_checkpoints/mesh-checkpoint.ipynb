```python
import drawsvg as draw
from perlin_numpy import generate_perlin_noise_3d

d = draw.Drawing(800, 800, origin='top-left', displayInline=False)

radius = 1
padding = 2
grid_divisions = 98
noise_scaler = 200.0
x_tick = d.height / grid_divisions
y_tick = d.width / grid_divisions

noise = generate_perlin_noise_3d(
    shape=[grid_divisions-padding, grid_divisions-padding, 2], 
    res=[4,4,2], 
)
# noise = (noise + 1.0) / 2.0

for x in range(padding,grid_divisions-padding):
    for y in range(padding,grid_divisions-padding):
        c = draw.Circle(
            x * x_tick + noise[x][y][0]*noise_scaler, 
            y * y_tick + noise[x][y][1]*noise_scaler, 
            radius,
            fill='black', 
            # fill_opacity=0.0,
            stroke_width=0, 
            stroke='black')
        d.append(c)
         
# Display in Jupyter notebook
d.rasterize()  
d
```

```python
import drawsvg as draw
from perlin_numpy import generate_perlin_noise_3d
import numpy as np

d = draw.Drawing(800, 800, origin='top-left', displayInline=False)

radius = 1
padding = 6
grid_divisions = 98
noise_scaler = 100.0
x_tick = d.height / grid_divisions
y_tick = d.width / grid_divisions

noise = generate_perlin_noise_3d(
    shape=[grid_divisions-padding, grid_divisions-padding, 2], 
    res=[4,4,2], 
    # tileable=[True,True,True]
)
# noise = (noise + 1.0) / 2.0

coords = []
for x in range(padding,grid_divisions-padding):
    row = []
    for y in range(padding,grid_divisions-padding):
        row.append((
            x * x_tick + noise[x][y][0]*noise_scaler,
            y * y_tick + noise[x][y][1]*noise_scaler))
    coords.append(row)
    
for x in range(grid_divisions-2*padding):
    current_line = []
    for y in range(grid_divisions-2*padding):
        current_line.append(coords[x][y][0])
        current_line.append(coords[x][y][1])
    c = draw.Lines(
        *current_line,
        fill='red', 
        fill_opacity=0.0,
        stroke_width=1, 
        stroke='black')
    d.append(c)
    
for y in range(grid_divisions-2*padding):
    current_line = []
    for x in range(grid_divisions-2*padding):
        current_line.append(coords[x][y][0])
        current_line.append(coords[x][y][1])
    c = draw.Lines(
        *current_line,
        fill='red', 
        fill_opacity=0.0,
        stroke_width=1, 
        stroke='black')
    d.append(c)
    
# Display in Jupyter notebook
d.rasterize()  
d
```
